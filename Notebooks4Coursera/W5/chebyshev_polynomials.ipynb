# Chebyshev Expansion of Cosine Function by Regression

$f(x)=cos(nx)$ where $n=1, 2, 3, ...$

$f(x)=cos(nx)$ is expanded using Chebyshev Polynomials into form of $a \cdot cos^n(x) + b \cdot cos^{n-1}(x) + c \cdot cos^{n-2}(x) + ... + z \cdot cos^0(x)$ 

Where $a, b, c, ..., z$ are the Chebyshev coefficients

For example:

$f(x)=cos(2x)=a \cdot cos^2(x) + b \cdot cos(x) + c$ <br>
$f(x)=cos(3x)=a \cdot cos^3(x) + b \cdot cos^2(x) + c \cdot cos(x) + d$ <br>
$f(x)=cos(4x)=a \cdot cos^4(x) + b \cdot cos^3(x) + c \cdot cos^2(x) + d \cdot cos(x) + e$
$f(x)=cos(5x)=a \cdot cos^5(x) + b \cdot cos^4(x) + c \cdot cos^3(x) + d \cdot cos^2(x) + e \cdot cos(x) + f$

The coefficients can be determined using regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y1 = np.cos(x)
y2 = np.cos(2*x)
y3 = np.cos(3*x)
y4 = np.cos(4*x)
y5 = np.cos(5*x)

import matplotlib.ticker as tck # needed to print the x axis as ...*pi

plt.figure(figsize=(10,5))

p1 = plt.plot(x/np.pi, y1)
p2 = plt.plot(x/np.pi, y2)
p3 = plt.plot(x/np.pi, y3)
p4 = plt.plot(x/np.pi, y4)
p5 = plt.plot(x/np.pi, y5)

plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0]), (['f(x)=sin(x)', 'f(x)=sin(2x)', 'f(x)=sin(3x)', 'f(x)=sin(4x)', 'f(x)=sin(5x)']))
plt.gca().xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.grid('true')
plt.xlim(0, 2)
plt.show()
```

## Function $f(x)=cos(2x)$

```python
# real function 
x = np.linspace(0, 2*np.pi, 1000)
y2 = np.cos(2*x)

# chebyshev expansion ax^2 + bx + c
def y2_cheby(x, a, b, c):
    cos = np.cos(x)
    y = (a * cos**2) + (b * cos) + c
    return y

# non-linear regression fit

from scipy.optimize import curve_fit

popt, pcov = curve_fit(y2_cheby, x, y2)
print('Chebyshev coefficients:', popt)

# plot real function and chebyshev 
y2_ch = (popt[0] * np.cos(x)**2) + (popt[1] * np.cos(x)) + popt[2]

import matplotlib.ticker as tck # needed to print the x axis as ...*pi

plt.figure(figsize=(10,5))
plt.plot(x/np.pi, y2, '.', color='red')
plt.plot(x/np.pi, y2_ch, color='black')
plt.gca().xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.grid('true')
plt.xlim(0, 2)
plt.show()
```

## Function $f(x)=cos(3x)$

```python
# real function 
x = np.linspace(0, 2*np.pi, 1000)
y3 = np.cos(3*x)

# chebyshev expansion ax^3 + bx^2 + cx + d
def y3_cheby(x, a, b, c, d):
    cos = np.cos(x)
    y = (a * cos**3) + (b * cos**2) + (c * cos) + d
    return y

# non-linear regression fit

from scipy.optimize import curve_fit

popt, pcov = curve_fit(y3_cheby, x, y3)
print('Chebyshev coefficients:', popt)

# plot real function and chebyshev 
y3_ch = (popt[0] * np.cos(x)**3) + (popt[1] * np.cos(x)**2) + (popt[2] * np.cos(x)) + popt[3]

import matplotlib.ticker as tck # needed to print the x axis as ...*pi

plt.figure(figsize=(10,5))
plt.plot(x/np.pi, y3, '.', color='red')
plt.plot(x/np.pi, y3_ch, color='black')
plt.gca().xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.grid('true')
plt.xlim(0, 2)
plt.show()
```
