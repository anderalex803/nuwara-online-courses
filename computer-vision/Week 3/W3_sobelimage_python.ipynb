```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```
!wget http://people.math.sc.edu/Burkardt/data/tif/cameraman.tif
```

```
filename = '/content/cameraman.tif'

cameraman = cv2.imread(filename)
cv2_imshow(cameraman)
```

```
# calculation laplacian, sobel horizontal, and sobel vertical 

laplacian = cv2.Laplacian(cameraman,cv2.CV_64F)
Gx = cv2.Sobel(cameraman,cv2.CV_64F,1,0,ksize=1)
Gy = cv2.Sobel(cameraman,cv2.CV_64F,0,1,ksize=1)

plt.figure(figsize=(10,10))

plt.subplot(2,2,1),plt.imshow(cameraman)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian)
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(Gx)
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(Gy)
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```
# Find magnitude and direction
Gmag = cv2.magnitude(Gx, Gy)
Gdir = np.arctan2(Gy, Gx) * (180 / np.pi)

plt.figure(figsize=(10,10))

plt.subplot(2,2,1),plt.imshow(cameraman)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(Gmag)
plt.title('Sobel Magnitude'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(Gdir)
plt.title('Sobel Direction'), plt.xticks([]), plt.yticks([])

plt.show()
```
